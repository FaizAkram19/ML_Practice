```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('salary_data.csv')
x_train=df['YearsExperience'].values
y_train=df['Salary'].values
```

```python
from sklearn.linear_model import LinearRegression

# 1. Reshape data
# Scikit-learn expects 'X' to be a 2D array (rows, columns). 
# Your x_train is 1D (rows,), so we reshape it to (rows, 1).
X = x_train.reshape(-1, 1) 
y = y_train

# 2. Create the model
model = LinearRegression()

# 3. Train the model (This does the Gradient Descent for you!)
model.fit(X, y)

# 4. Get the results
print(f"Scikit-Learn w (coef): {model.coef_}")
print(f"Scikit-Learn b (intercept): {model.intercept_}")

# 5. Predict
prediction = model.predict([[12]])
print(f"Scikit-Learn Prediction for 12 years: {prediction}")
```

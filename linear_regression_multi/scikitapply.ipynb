```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# 1. Setup the data (Same as before)
df = pd.read_csv('housing_data.csv')
x_train = df[['Size', 'Bedrooms', 'Age']].values
y_train = df['Price'].values

# 2. Scale the features (Replaces your 'zscore' function)
# StandardScaler automatically computes mean and std (fit) and scales the data (transform)
scaler = StandardScaler()
x_train_norm = scaler.fit_transform(x_train)

# 3. Train the model (Replaces 'compute_cost', 'gradient_descent', etc.)
# The .fit() method runs the optimization algorithm in the background
sgdr = LinearRegression()
sgdr.fit(x_train_norm, y_train)

# 4. View Parameters
print(f"Weights (w): {sgdr.coef_}")
print(f"Bias (b): {sgdr.intercept_}")

# 5. Make a Prediction (Replaces the manual dot product steps)
# We must scale the new input using the SAME scaler we just trained
x_house = np.array([[2.0, 3, 15]]) 
x_house_norm = scaler.transform(x_house)

prediction = sgdr.predict(x_house_norm)
print(f"Predicted Price: ${prediction[0]:.2f}k")
```
